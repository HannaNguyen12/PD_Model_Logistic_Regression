### 3.1 Preprocess Categorical Data

#### * Help function to Calculate Weight of Information (WoE)

```python
def woe_vi_discrete(independent, dependent, adj=0.5):
    df = pd.concat([independent, dependent], axis=1)
    target = df.columns[1]
    feature = df.columns[0]
    res = df.groupby(feature)[target].agg(total_cases='count', default_cases='sum').reset_index()
    res['non_default_cases'] = res['total_cases'] - res['default_cases']
    # Apply Smoothing: Add 'adj' (usually 0.5) to avoid division by zero or log(0)
    res['prop_non_default'] = (res['non_default_cases'] + adj) / (res['non_default_cases'].sum() + 2 * adj)
    res['prop_default'] = (res['default_cases'] + adj) / (res['default_cases'].sum() + 2 * adj)
    prop_non_default = (res['non_default_cases'] + adj) / (res['non_default_cases'].sum() + 2 * adj)
    prop_default = (res['default_cases'] + adj) / (res['default_cases'].sum() + 2 * adj)
    res['woe'] = np.log(prop_non_default / prop_default)
    res['iv_contribution'] = (prop_non_default - prop_default) * res['woe']
    total_iv = res['iv_contribution'].sum() 
    res['total_iv'] = total_iv
    return res.sort_values(by='woe', ascending=False)
```

```python
def get_numeric_iv_report(X, y, numeric_cols, bins=10):
    iv_results = []
    
    for col in numeric_cols:
        df_temp = pd.DataFrame({'feature': X[col], 'target': y})
        df_temp['feature'] = df_temp['feature'].fillna(df_temp['feature'].median())
        if df_temp.empty:
            continue
        try:
            df_temp['bin'] = pd.qcut(df_temp['feature'], q=bins, labels = False, duplicates = 'drop')
            stats = df_temp.groupby('bin')['target'].agg(['count', 'sum'])
            stats.columns = ['total', 'defaults']
            stats['non_defaults'] = stats['total'] - stats['defaults']
            adj = 0.5
            prop_non_default = (stats['non_defaults'] + adj) / (stats['non_defaults'].sum() + 2 * adj)
            prop_default = (stats['defaults'] + adj) / (stats['defaults'].sum() + 2 * adj)
            woe = np.log(prop_non_default / prop_default)
            iv_contribution = (prop_non_default - prop_default) * woe
            total_iv = iv_contribution.sum()
            iv_results.append({'Feature': col, 'IV': total_iv})
        except Exception as e:
            print(f"Could not calculate IV for {col}: {e}")
    return pd.DataFrame(iv_results).sort_values(by='IV', ascending=False)

# Execute:
pd.options.display.float_format = '{:.4f}'.format
```

#### Binning numeric Features

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

def bin_one_variable(
    X_train,
    y_train,
    X_test,
    col,
    method='tree',   # 'tree' or 'quantile'
    max_bins=6,
    q_bins=8,
    sample_size=200000,
    random_state=42
):
    
    # copy to avoid modifying original
    X_train_out = X_train.copy()
    X_test_out = X_test.copy()
    
    # sample for speed
    sample = X_train[[col]].copy()
    sample['target'] = y_train
    sample = sample.sample(min(sample_size, len(sample)), random_state=random_state)
    
    mask = sample[col].notna()
    
    # -------------------
    # TREE BINNING
    # -------------------
    if method == 'tree':
        clf = DecisionTreeClassifier(
            max_leaf_nodes=max_bins,
            min_samples_leaf=0.05,
            random_state=random_state
        )
        
        clf.fit(sample.loc[mask, [col]], sample.loc[mask, 'target'])
        
        thresholds = clf.tree_.threshold
        thresholds = thresholds[thresholds != -2]
        thresholds = np.sort(thresholds)
    
    # -------------------
    # QUANTILE BINNING
    # -------------------
    elif method == 'quantile':
        quantiles = np.linspace(0, 1, q_bins + 1)
        thresholds = np.quantile(sample.loc[mask, col], quantiles)
        thresholds = np.unique(thresholds)[1:-1]
    
    else:
        raise ValueError("method must be 'tree' or 'quantile'")
    
    # apply binning
    X_train_out[col + '_bin'] = np.searchsorted(thresholds, X_train[col])
    X_test_out[col + '_bin'] = np.searchsorted(thresholds, X_test[col])
    
    return X_train_out, X_test_out, thresholds
```
